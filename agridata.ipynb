```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import plotly.express as px
```

```python
#Question-1-Top 7 Rice Producing States:
# SELECT
top7_rice = df[df['rice_production_(1000_tons)'] > 0] \
    .groupby('state_name')['rice_production_(1000_tons)'] \
    .sum().sort_values(ascending=False).head(7)
print(top7_rice)

# PLOT
top7_rice.plot(kind='bar', title='Top 7 Rice Producing States',color='orange')
plt.ylabel('Production (1000 tons)')
plt.xticks(rotation=45)
plt.show()
```

```python
#Question-2-Top 5 Wheat Producing States:
# SELECT
top5_wheat = df[df['wheat_production_(1000_tons)'] > 0] \
    .groupby('state_name')['wheat_production_(1000_tons)'] \
    .sum().sort_values(ascending=False).head(5)
print(top5_wheat)


# PLOT
top5_wheat.plot(kind='bar', title='Top 5 Wheat Producing States',color='yellow',width=0.5,)
plt.ylabel('Production (1000 tons)')
plt.xticks(rotation=45)
plt.show()
```

```python
#Question-3-Oilseed Production by Top 5 States:
# SELECT
top5_oilseed = df[df['oilseeds_production_(1000_tons)'] > 0] \
    .groupby('state_name')['oilseeds_production_(1000_tons)'] \
    .sum().sort_values(ascending=False).head(5)
print(top5_oilseed)
# PLOT
top5_oilseed.plot(kind='bar', title='Top 5 Oilseed Producing States')
plt.ylabel('Production (1000 tons)')
plt.xticks(rotation=45)
plt.show()
```

```python
#Question-4-Top 7 Sunflower Producing States:
# SELECT
top7_sunflower = df[df['sunflower_production_(1000_tons)'] > 0] \
    .groupby('state_name')['sunflower_production_(1000_tons)'] \
    .sum().sort_values(ascending=False).head(7)
print(top7_sunflower)

# PLOT
top7_sunflower.plot(kind='bar', title='Top 7 Sunflower Producing States')
plt.ylabel('Production (1000 tons)')
plt.xticks(rotation=45)
plt.show()
```

```python
#Question-5-Sugarcane Production Over Last 50 Years:
# SELECT
sugarcane_trend = df.groupby('year')['sugarcane_production_(1000_tons)'] \
    .sum().sort_index()
print(sugarcane_trend)

# PLOT
sugarcane_trend.plot(kind='line', title='Sugarcane Production Over Last 50 Years')
plt.ylabel('Production (1000 tons)')
plt.xlabel('Year')
plt.show()
```

```python
#Question-6-Rice vs Wheat Production (Last 50 Years):
# SELECT
rice_wheat = df.groupby('year')[['rice_production_(1000_tons)', 'wheat_production_(1000_tons)']] \
    .sum().sort_index()
print(rice_wheat)

# PLOT
rice_wheat.plot(kind='line', title='Rice vs Wheat Production (Last 50 Years)')
plt.ylabel('Production (1000 tons)')
plt.xlabel('Year')
plt.show()
```

```python
#Question-7-Rice Production by West Bengal Districts:
# SELECT
wb_rice = df[(df['state_name'] == 'West Bengal') & (df['rice_production_(1000_tons)'] > 0)] \
    .groupby('dist_name')['rice_production_(1000_tons)'] \
    .sum().sort_values(ascending=False)
print(wb_rice)

# PLOT
wb_rice.plot(kind='bar', title='Rice Production by Districts in West Bengal')
plt.ylabel('Production (1000 tons)')
plt.xticks(rotation=90)
plt.show()
```

```python
#QUESTION-1-Year-wise Trend of Rice Production Across States (Top 3)
mycursor.execute("""select year, state_name, sum(`rice_production_(1000_tons)`) as total_rice_production
from agridata.CROP_DETAILS
where state_name in (select state_name from agridata.CROP_DETAILS group by state_name order by sum(`rice_production_(1000_tons)`) desc limit 3)
group by year, state_name
order by year desc, total_rice_production desc""")
out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))
```

```python
#QUESTION-2-Top 5 Districts by Wheat Yield Increase Over the Last 5 Years
mycursor.execute("""select dist_name, (max(`wheat_yield_(kg_per_ha)`)-min(`wheat_yield_(kg_per_ha)`)) as wheat_yield_increase
from agridata.CROP_DETAILS
where year>=(select max(year) from agridata.CROP_DETAILS)-4
group by dist_name
order by wheat_yield_increase desc limit 5""")
out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))
```

```python
#QUESTION-3-States with the Highest Growth in Oilseed Production (5-Year Growth Rate)
mycursor.execute("""with oilseed_growth as (select
state_name,
min(case when year = 2013 then `oilseeds_production_(1000_tons)` end) as prod_2013,
max(CASE WHEN year = 2017 THEN `oilseeds_production_(1000_tons)` end) as prod_2017
from agridata.CROP_DETAILS
group by state_name)
select state_name, prod_2013, prod_2017,
((prod_2017 - prod_2013) * 100.0 / nullif(prod_2013, 0)) as growth_rate
from oilseed_growth
order by growth_rate desc limit 3""")
out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))
```

```python
#QUESTION-4-District-wise Correlation Between Area and Production for Major Crops (Rice, Wheat, and Maize)
mycursor.execute("""select dist_name, 'rice' as crop, sum(`rice_production_(1000_tons)`) as production,
sum(`rice_area_(1000_ha)`) as area
from agridata.CROP_DETAILS
group by dist_name
union all
select dist_name, 'wheat', sum(`wheat_production_(1000_tons)`), sum(`wheat_area_(1000_ha)`)
from agridata.CROP_DETAILS
group by dist_name
union all
select dist_name,'maize', sum(`maize_production_(1000_tons)`), sum(`maize_area_(1000_ha)`)
from agridata.CROP_DETAILS
group by dist_name""")
out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))
```

```python
#QUESTION-5-Yearly Production Growth of Cotton in Top 5 Cotton Producing States
mycursor.execute("""select year, state_name,total_cotton_production, prev_year_production,
round((total_cotton_production - prev_year_production) / prev_year_production * 100, 2) as growth_percentage
from (select year, state_name,sum(`cotton_production_(1000_tons)`) as total_cotton_production,
lag(sum(`cotton_production_(1000_tons)`)) over (partition by state_name order by year) as prev_year_production
from agridata.CROP_DETAILS
where state_name in (select state_name
from agridata.CROP_DETAILS
group by state_name
order by sum(`cotton_production_(1000_tons)`) desc
limit 5)
group by year, state_name) as subquery
order by year desc, total_cotton_production desc""")
out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))
```

```python
#QUESTION-6-Districts with the Highest Groundnut Production in 2020
mycursor.execute("""select dist_name, sum(`groundnut_production_(1000_tons)`) as total_groundnut_production
from agridata.CROP_DETAILS
where year=2020
group by dist_name
order by total_groundnut_production desc""")
out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))
```

```python
#QUESTION-10-Compare the Production of Wheat and Rice for the Top 5 States Over 10 Years
mycursor.execute("""SELECT state_name, year, SUM(`wheat_production_(1000_tons)`) AS total_wheat_production,
SUM(`rice_production_(1000_tons)`) AS total_rice_production
FROM agridata.CROP_DETAILS
WHERE state_name IN (SELECT state_name
FROM agridata.CROP_DETAILS
WHERE year BETWEEN (SELECT MAX(year) - 9 FROM agridata.CROP_DETAILS) AND (SELECT MAX(year) FROM agridata.CROP_DETAILS)
GROUP BY state_name
ORDER BY SUM(`wheat_production_(1000_tons)`) + SUM(`rice_production_(1000_tons)`) DESC
LIMIT 5)
AND year BETWEEN (SELECT MAX(year) - 9 FROM agridata.CROP_DETAILS) AND (SELECT MAX(year) FROM agridata.CROP_DETAILS)
GROUP BY state_name, year
ORDER BY state_name, year DESC""")
out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))
```
